##Einstein E=mc²
This program prompts the user for mass as integer (kg) and then outputs the equivalent number fo Joules as an integer.

```python
def main():
  # Prompt the user for mass as an integer
  mass_kg = int(input("Enter mass in kilograms (kg): "))

  # Define the speed of light (c)
  speed_of_light = 300000000  # Approximately 3 x 10^8 m/s

  # Calculate energy using E = mc²
  energy_joules = mass_kg * (speed_of_light ** 2)

  # Output the equivalent number of Joules as an integer
  print(f"Equivalent energy in Joules: {energy_joules}")

main()
```

##Tip Calculator
This program will allow you to input how much a bill is and percentage you would like to leave for a tip and it will do the math for you.

```python
def main():
    dollars = dollars_to_float(input("How much was the meal? "))
    percent = percent_to_float(input("What percentage would you like to tip? "))
    tip = dollars * percent
    print(f"Leave ${tip:.2f}")

def dollars_to_float(d):
    # Remove the dollar sign and convert to float
    return float(d.replace("$", ""))

def percent_to_float(p):
    # Remove the percentage sign, convert to float, and divide by 100
    return float(p.replace("%", "")) / 100

main()
```

```python
# if you want to execute a SQL command you must start with the following
import sqlite3
import os

# Define the database file path using the correct path provided by the user
db_file = '/content/drive/MyDrive/Colab Notebooks/SuperDuper/cyberchase/cyberchase.db'

# Connect to the SQLite database
conn = None
try:
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    print(f"Successfully connected to the database: {db_file}")

    # Query all tables in the database
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    print("\nTables in the database:")
    if tables:
        for table in tables:
            print(table[0])
    else:
        print("No tables found in the database.")

except sqlite3.Error as e:
    print(f"Error connecting to or querying database: {e}")

finally:
    # Close the connection
    if conn:
        conn.close()
        print("Database connection closed.")
```

```python
#This cell will return all data from the episodes table
import sqlite3
import os

# Define the database file path - using the path confirmed in the previous step
db_file = '/content/drive/MyDrive/Colab Notebooks/SuperDuper/cyberchase/cyberchase.db'

# Connect to the SQLite database
conn = None
try:
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    print(f"Querying data from the episodes table in {db_file}:")

    # Execute the SELECT * query on the episodes table
    cursor.execute("SELECT * FROM episodes;")

    # Fetch all the results
    results = cursor.fetchall()

    # Print the results
    if results:
        print("\nData in episodes table:")
        for row in results:
            print(row)
    else:
        print("No data found in episodes table.")

except sqlite3.Error as e:
    print(f"Error querying database: {e}")

finally:
    # Close the connection
    if conn:
        conn.close()
        print("Database connection closed.")
```

###Visualizing Data
Let's start with a simple bar graph. Since we know that we can get the number of episodes in a season and we can get how many seasons the show had.

Task:
Visualize the number of episodes per season from the `episodes` table in the SQLite database "/content/drive/MyDrive/Colab Notebooks/SuperDuper/cyberchase/cyberchase.db" using a bar graph generated with matplotlib.

```python
import sqlite3
import os

db_file = '/content/drive/MyDrive/Colab Notebooks/SuperDuper/cyberchase/cyberchase.db'

conn = None
episode_seasons = []
try:
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()
    cursor.execute("SELECT season FROM episodes;")
    episode_seasons = cursor.fetchall()
except sqlite3.Error as e:
    print(f"Error querying database: {e}")
finally:
    if conn:
        conn.close()
```

## Process data

### Subtask:
Load the queried data into a pandas DataFrame and count the number of episodes for each season.


**Reasoning**:
The previous step successfully retrieved the season data into the `episode_seasons` list. Now, as per the instructions, I will load this data into a pandas DataFrame, count the episodes per season, and prepare the data for plotting.



```python
import pandas as pd

# Convert the list of tuples into a pandas DataFrame
df_seasons = pd.DataFrame(episode_seasons, columns=['season'])

# Count the number of episodes per season
episodes_per_season = df_seasons['season'].value_counts().sort_index()

# Display the result
print("Number of episodes per season:")
print(episodes_per_season)
```

## Visualize data

### Subtask:
Create a bar graph using matplotlib to show the number of episodes per season.


**Reasoning**:
Generate a bar plot using matplotlib to visualize the number of episodes per season.



```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 6))
episodes_per_season.plot(kind='bar')

# Add title and labels
plt.title("Number of Episodes per Season")
plt.xlabel("Season")
plt.ylabel("Number of Episodes")

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Display the plot
plt.tight_layout()
plt.show()
```
